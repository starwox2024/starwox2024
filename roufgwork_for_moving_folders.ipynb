```python
import requests
from PIL import Image
from io import BytesIO

def retrieve_image_from_uri(image_uri):
    # Send a GET request to the URI
    response = requests.get(image_uri)

    # Check if the request was successful
    if response.status_code == 200:
        # Load the image data into a Pillow Image object
        image = Image.open(BytesIO(response.content))
        return image
    else:
        raise Exception(f"Failed to retrieve image. Status code: {response.status_code}")

# Example usage

image_uri = 'https://generativelanguage.googleapis.com/v1beta/files/z4dpyualdzs0'
image = retrieve_image_from_uri(image_uri)

# Display the image
image.show()

# Optionally, save the image to a file
#image.save("downloaded_image.jpg")
```
